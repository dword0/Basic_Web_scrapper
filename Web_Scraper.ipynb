##Building a simple web scraper 

```python
# First, we need to import all the libraries that we are going to use.
import urllib.request
from bs4 import BeautifulSoup
```

```python
# Next, declare a variable for the url of the page
quote_page = 'https://coinmarketcap.com/currencies/dogecoin/'
```

```python
# Now make use of the Python urllib.request to get the HTML page of the url declared.
page = urllib.request.urlopen(quote_page).read()

# Just trying to find the return type of the function
print(type(urllib.request.urlopen(quote_page).read()))
```

```python
# Printing the content extracted
print(page)

# Parse the page into BeautifulSoup format so we can use BeautifulSoup and store it in a variabel soup to work on it.
soup = BeautifulSoup(page, 'html.parser')
```

```python
# BeautifulSoup can help us get into these layers and extract the content with find(). In this case, since the HTML class name is unique on this page, we can simply query <div class="sc-1q9q90x-0 iYFMbU h1___3QSYG">
namebox = soup.find('h2',attrs={'class':'sc-1q9q90x-0 iYFMbU h1___3QSYG'})

#After we have the tag, we can get the data by getting its text.
cr_name = namebox.text.strip()
```

```python
# Repeat the above steps to find the price of the coin
price = soup.find('div',attrs={'class':'priceValue___11gHJ'})
cr_price = price.text.strip()
```

```python
print(cr_name," = ",cr_price)
```

For dynamic websites we will have to use selenium. The above code can only be used for static websites
